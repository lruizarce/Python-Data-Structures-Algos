```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_dll(self):
        curr = self.head
        while curr:
            print(curr.value, "--->", end='')
            curr = curr.next

    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            # points to the new_node
            self.tail.next = new_node
            # new node points to tail
            new_node.prev = self.tail
            # tail points to new_node
            self.tail = new_node

        self.length +=1
        return True

    def pop(self):
        if self.length ==0:
            return None
        curr = self.tail
        if self.length ==1:
            self.head = None
            self.tail = None
        else:
            # tail points to previous node
            self.tail = self.tail.prev
            # detaches from LL
            self.tail.next = None
            # detaches from LL
            curr.prev = None
        self.length -=1
        return curr

    def prepend(self, value):
        new_node = Node(value)
        if self.length ==0:
            self.head = new_node
            self.tail = new_node
        else:
            # points to the head
            new_node.next = self.head
            # head.prev points to new_node
            self.head.prev = new_node
            # head points to new_node
            self.head = new_node
        self.length +=1
        return True

    def pop_first(self):
        if self.length ==0:
            return None
        curr = self.head
        if self.length ==1:
            self.head = None
            self.tail = None
        else:
            # head points to the next node
            self.head = self.head.next
            # detaches head from LL
            self.head.prev = None
            # garbage 
            curr.next = None

        self.length -=1
        return curr
    
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        curr = self.head
        # optimize search
        if index < self.length/2:
            for _ in range(index):
                curr = curr.next
            return curr
        else:
            curr = self.tail
            for _ in range(self.length -1, index, -1):
                curr = curr.prev
            return curr
            
    def setter(self, index, value):
        curr = self.get(index)
        if curr:
            curr.value = value
            return True
        return False
    
    def insert(self, value, index):
        if index < 0 | index > self.length:
            return False
        
        if index == 0:
            return self.prepend(value)
        
        if index == self.length:
            return self.append(value)
        else:
            new_node = Node(value)
            before = self.get(index -1)
            after = self.head
            new_node.prev = before
            new_node.next = after
            before.next = new_node
            after.prev = new_node
        self.length +=1
        return True
    
    def remove(self, index):
        if index < 0 | index <= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length -1:
            return self.pop()
        
        curr = self.get(index)
        curr.next.prev = curr.prev
        curr.prev.next = curr.next
        curr.next = None
        curr.prev = None
        self.length+=1
        return curr    
            
```
